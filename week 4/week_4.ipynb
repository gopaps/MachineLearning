```python
#Data manipulation libraries
import numpy as np
import pandas as pd

#Data Visulaization Libraries
import matplotlib.pyplot as plt
import seaborn as sns
#To ensure that plots are displayed inline in jupyter notebook
%matplotlib inline
```

```python
#Memuat dataset dari Google Drive
import pandas as pd
#Path ke file dataset di Google Drive
file_path='/content/drive/My Drive/dataset/StudentsPerformance.csv'
#Memuat dataset ke dalam DataFrame pandas
df = pd.read_csv(file_path)
#Menampilkan beberapa baris pertama dari dataset
df.head()
```

```python
#Melihat shape dari DataFrame
print(df.shape)
```

```python
# Melihat statistical summary dari DataFrame
print(df.describe())
```

```python
print(df.info())
```

```python

# Set the plot style
sns.set(style="whitegrid")
```

```python
# 1. Visualisasi Distribusi Gender
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="gender", palette="pastel")
plt.title('Distribution of Gender in Dataset')
plt.show()
```

```python
# 2. Visualisasi Distribusi Nilai Matematika
plt.figure(figsize=(8,5))
sns.histplot(df['math score'], kde=True, color='skyblue')
plt.title('Distribution of Math Scores')
plt.show()
```

```python
# 3. Visualisasi Distribusi Nilai Membaca
plt.figure(figsize=(8,5))
sns.histplot(df['reading score'], kde=True, color='lightgreen')
plt.title('Distribution of Reading Scores')
plt.show()
```

```python
# 4. Visualisasi Distribusi Nilai Menulis
plt.figure(figsize=(8,5))
sns.histplot(df['writing score'], kde=True, color='salmon')
plt.title('Distribution of Writing Scores')
plt.show()
```

```python
# 5. Visualisasi Performa Berdasarkan Race/Ethnicity
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="race/ethnicity", y="math score", palette="Set3")
plt.title('Math Score by Race/Ethnicity')
plt.show()
```

```python
# 6. Visualisasi Performa Berdasarkan Parental Level of Education
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="parental level of education", y="math score", palette="Set2")
plt.title('Math Score by Parental Level of Education')
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. Visualisasi Pengaruh Kursus Persiapan Tes Terhadap Skor Matematika
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="test preparation course", y="math score", palette="muted")
plt.title('Math Score by Test Preparation Course')
plt.show()
```

```python
# Rename kolom dataset
df.rename(columns={
    'gender': 'gender',
    'race/ethnicity': 'race/ethnicity',
    'parental level of education': 'parental level of education',
    'lunch': 'lunch',
    'test preparation course': 'test preparation course',
    'math score': 'math score',
    'reading score': 'reading score',
    'writing score': 'writing score'
}, inplace=True)

# Menampilkan nama kolom setelah diubah
print(df.columns)
```

```python
# Klasifikasi kolom berdasarkan tipe data
categorical_columns = df.select_dtypes(include=['object']).columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Menampilkan hasil klasifikasi kolom
print("Categorical Columns:")
print(categorical_columns)

print("\nNumerical Columns:")
print(numerical_columns)
```

```python
#Memerikasa apakah ada null di dataset
df.isnull().sum()
```

```python

# Menghitung total skor dari Math, Reading, dan Writing untuk setiap siswa
df['total score'] = df['math score'] + df['reading score'] + df['writing score']
```

```python
# 1. Rata-rata Total Skor Berdasarkan Gender
plt.figure(figsize=(8,6))
sns.barplot(data=df, x="gender", y="total score", palette="Set2")
plt.title('Average Total Score by Gender')
plt.show()
```

```python
# 2. Rata-rata Total Skor Berdasarkan Race/Ethnicity
plt.figure(figsize=(8,6))
sns.barplot(data=df, x="race/ethnicity", y="total score", palette="Set3")
plt.title('Average Total Score by Race/Ethnicity')
plt.show()
```

```python
# 3. Rata-rata Total Skor Berdasarkan Parental Level of Education
plt.figure(figsize=(10,6))
sns.barplot(data=df, x="parental level of education", y="total score", palette="muted")
plt.title('Average Total Score by Parental Level of Education')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Rata-rata Total Skor Berdasarkan Test Preparation Course
plt.figure(figsize=(8,6))
sns.barplot(data=df, x="test preparation course", y="total score", palette="Set1")
plt.title('Average Total Score by Test Preparation Course')
plt.show()
```
