Teknik seperti Bag-of-Words dan TF-IDF (Term Frequency-Inverse Document Frequency) digunakan untuk merepresentasikan teks dalam bentuk numerik, sehingga dapat diproses oleh algoritma machine learning. Contoh implementasi menggunakan CountVectorizer dan TfidfVectorizer menunjukkan bagaimana teks dapat diubah menjadi vektor fitur yang relevan untuk klasifikasi.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report
from sklearn.datasets import fetch_20newsgroups
import torch
```

```python
# Periksa apakah GPU tersedia
if torch.cuda.is_available():
    device = torch.device("cuda")
    print("GPU tersedia. Menggunakan:", torch.cuda.get_device_name(0))
else:
    device = torch.device("cpu")
    print("GPU tidak tersedia. Menggunakan CPU.")
```

```python
# Memuat dataset teks (20 Newsgroups)
newsgroups_train = fetch_20newsgroups(subset='train', categories=['rec.sport.baseball', 'sci.space'], remove=('headers', 'footers', 'quotes'))
newsgroups_test = fetch_20newsgroups(subset='test', categories=['rec.sport.baseball', 'sci.space'], remove=('headers', 'footers', 'quotes'))

# Data dan label
X_train, y_train = newsgroups_train.data, newsgroups_train.target
X_test, y_test = newsgroups_test.data, newsgroups_test.target
```

```python
# Membuat Pipeline untuk Bag-of-Words dan Naive Bayes
pipeline = Pipeline([
    ('vectorizer', CountVectorizer(stop_words='english', ngram_range=(1, 2))),
    ('classifier', MultinomialNB())
])
```

```python
# Melatih model
pipeline.fit(X_train, y_train)

# Evaluasi model
y_pred = pipeline.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Akurasi Model: {accuracy * 100:.2f}%")
```

```python
# Laporan Klasifikasi
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=newsgroups_train.target_names))
```

```python
# Visualisasi fitur penting dengan TF-IDF
tfidf_pipeline = Pipeline([
    ('vectorizer', TfidfVectorizer(stop_words='english', ngram_range=(1, 2))),
    ('classifier', MultinomialNB())
])
```

```python
# Melatih ulang dengan TF-IDF
tfidf_pipeline.fit(X_train, y_train)

y_pred_tfidf = tfidf_pipeline.predict(X_test)
tfidf_accuracy = accuracy_score(y_test, y_pred_tfidf)
print(f"Akurasi Model dengan TF-IDF: {tfidf_accuracy * 100:.2f}%")
```

```python
# Mengambil fitur penting
vectorizer = tfidf_pipeline.named_steps['vectorizer']
classifier = tfidf_pipeline.named_steps['classifier']

feature_names = vectorizer.get_feature_names_out()
important_features = np.argsort(classifier.feature_log_prob_[0])[-10:]
print("10 Fitur Paling Penting:")
for i in important_features:
    print(feature_names[i])
```

Pada bab ini, model Multinomial Naive Bayes digunakan untuk mengklasifikasikan dataset 20 Newsgroups yang terdiri dari kategori seperti olahraga dan sains. Model ini dievaluasi dengan metrik akurasi dan laporan klasifikasi, memberikan wawasan tentang performa pada data uji. Selain itu, fitur-fitur penting dalam model diidentifikasi menggunakan nilai probabilitas logaritmik dari Naive Bayes, memberikan pemahaman yang lebih mendalam tentang pengaruh fitur tertentu dalam klasifikasi.

Bab ini menekankan pentingnya representasi data dalam NLP, karena hasil analisis sangat bergantung pada cara data diproses.
